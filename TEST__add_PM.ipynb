```python
import os
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/gdrive/MyDrive/reg_df')
PM_data = pd.read_csv('/content/gdrive/MyDrive/기상기후 데이터/data_PM.csv')
```

# test

## data select column

```python
test_df = df[['측정일시','시간','습도','일조','일사','시정','남중고도','PM10','target']]
```

```python
test_df
```

```python
test_df.isnull().sum()
```

```python
test_df = test_df.dropna(subset=['일조','일사','PM10'], how='any')
test_df.shape
```

```python
test_df.reset_index(inplace=True,drop=True)
test_df
```

## Particulate_matter test 

```python
num_2019 = test_df[test_df['측정일시'].astype(str).str[:4] =='2019'].index
num_2019
```

```python
train = test_df.iloc[:num_2019[0]]
test = test_df.iloc[num_2019]
print(train.shape)
print(test.shape)
```

```python
print("PM10 단일 정확도 \n")

x_train = train[['PM10']]
y_train = train[['target']]
x_test = test[['PM10']]
y_test = test[['target']]

model = LinearRegression()
model.fit(x_train,y_train)

y_pred_train = model.predict(x_train)
r2_train = r2_score(y_train, y_pred_train)

# Test
y_test_pred = model.predict(x_test)
r2_test = r2_score(y_test, y_test_pred)

print("train data: %.6f, test data: %.6f \n " %(r2_train, r2_test))
```

```python
col_list = [col for col in test_df.columns if (col != 'target') & (col != '측정일시') & (col != '시간') & (col != 'PM10')]
col_list
```

```python
print("PM10 + 변수 정확도 \n")
for i in col_list:
  x_train = train[[i,'PM10']]
  y_train = train[['target']]
  x_test = test[[i,'PM10']]
  y_test = test[['target']]

  model = LinearRegression()
  model.fit(x_train,y_train)
  #train
  y_pred_train = model.predict(x_train)
  r2_train = r2_score(y_train, y_pred_train)

  # Test
  y_test_pred = model.predict(x_test)
  r2_test = r2_score(y_test, y_test_pred)

  print(i,"& PM10 train data: %.6f, test data: %.6f \n " %(r2_train, r2_test))
```

```python
print("단일 변수 정확도 \n")
for i in col_list:
  x_train = train[[i]]
  y_train = train[['target']]
  x_test = test[[i]]
  y_test = test[['target']]

  model = LinearRegression()
  model.fit(x_train,y_train)
  #train
  y_pred_train = model.predict(x_train)
  r2_train = r2_score(y_train, y_pred_train)

  # Test
  y_test_pred = model.predict(x_test)
  r2_test = r2_score(y_test, y_test_pred)
  
  print(i,"train data: %.6f, test data: %.6f \n " %(r2_train, r2_test))
```
