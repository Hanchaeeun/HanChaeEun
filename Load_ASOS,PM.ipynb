```python
import os
import numpy as np
import pandas as pd
import datetime
```

# Load

## ASOS Data

```python
path = '/content/gdrive/MyDrive/기상기후 데이터/목포_ASOSdata/'
file_list = os.listdir(path)

df = pd.DataFrame()
for i in file_list:
  print(i)
  ASOSdata = pd.read_csv(path+i, encoding='CP949')
  df = pd.concat([df, ASOSdata])
```

```python
df = df.sort_values('일시').reset_index(drop=True)
df
```

```python
df['시간'] = pd.to_datetime(df['일시'], format='%Y-%m-%d %H').dt.hour
df['일시'] = df['일시'].str[:10]
```

```python
col_list = df.columns
col_name = []

for i in col_list:
  name = i.split("(")[0]
  col_name.append(name)
col_name
```

```python
df.columns = col_name
print(df.shape)
df.head(2)
```

## Particulate Matter Data

```python
path='/content/gdrive/MyDrive/기상기후 데이터/PMdata/'
file_list = os.listdir(path)

Data_15 = pd.DataFrame()
for i in file_list[4:8]:
  data = pd.read_csv(path+i)
  data = data[data['지역'] == '전남 목포시']
  Data_15 = pd.concat([Data_15,data])
  file_list.remove(i)
  print(path+i)
```

```python
Data_16 = pd.DataFrame()
for i in file_list[:4]:
  data = pd.read_csv(path+i, encoding='CP949')
  data = data[data['지역'] == '전남 목포시']
  Data_16 = pd.concat([Data_16,data])
  file_list.remove(i)
  print(path+i)
```

```python
PMdata = pd.DataFrame()
for i in file_list:
  data = pd.read_excel(path+i)
  data = data[data['지역'] == '전남 목포시']
  PMdata = pd.concat([PMdata,data])
  print(path+i)
```

```python
PMdata.sort_values('측정일시', inplace=True)
PMdata.drop(columns='망', inplace=True)
PMdata.head()
```

```python
Data_15.head()
```

```python
Data_16.head()
```

```python
PMdata = pd.concat([Data_15, Data_16, PMdata])
PMdata.reset_index(inplace=True, drop=True)
PMdata
```

# Preprocessing

## Data extraction

```python
PMdata['측정소명'].unique()
```

```python
PMdata = PMdata[PMdata['측정소명'] == '용당동']
PMdata.shape
```

```python
PMdata = PMdata.loc[:,["측정일시","PM10","PM25"]]
PMdata.reset_index(inplace=True, drop=True)
PMdata
```

```python
PMdata['시간'] = PMdata['측정일시'].astype(str).str[-2:].astype(int) - 1
```
